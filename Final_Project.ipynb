```python
import pandas as pd
```

```python
df = pd.read_excel('titanic3.xls')
```

```python
df['survived'].mean()
```

```python
df.describe()
```

```python
# preprocess
df = df.replace({'female':0,'male':1})
df = df.drop('body', axis=1)
df = df.drop('home.dest', axis=1)
df[['boat', 'cabin']] = df[['boat', 'cabin']].fillna(0)
```

```python
# pclass
df_pclass_survived = df[['pclass','survived']]
df_pclass_survived
```

```python
res_pclass = df_pclass_survived.groupby('pclass').agg(['sum', 'count']).reset_index()
res_pclass
```

```python
# sex
df_sex_survived = df[['sex','survived']]
df_sex_survived
```

```python
res_sex = df_sex_survived.groupby('sex').agg(['sum', 'count']).reset_index()
res_sex # 0 for female and 1 for male
```

```python
# age
df_age_survived_dropna = df[['age','survived']].dropna()
df_age_survived_dropna # mean is 30
```

```python
def age_category(age):
    if age < 15:
        return '<15'
    elif 15 <= age <= 40:
        return '15-40'
    else:
        return '>40'

df_age_survived_dropna['age_group'] = df_age_survived_dropna['age'].apply(age_category)
res_age = df_age_survived_dropna.groupby('age_group')['survived'].agg(['sum', 'count']).reset_index()
res_age
```

```python
df_fare_survived = df[['fare','survived']] # mean fare is 33
df_fare_survived
```

```python
def fare_category(fare):
    if fare < 30:
        return '<30'
    elif 30 <= fare <= 100:
        return '30-100'
    else:
        return '>100'
df_fare_survived['fare_category'] = df_fare_survived['fare'].apply(fare_category)
res_fare = df_fare_survived.groupby('fare_category')['survived'].agg(['sum', 'count']).reset_index()
res_fare
```

```python
df_boat_survived = df[['boat','survived']]
df_boat_survived
```

```python
res_boat = df_boat_survived.groupby('boat').agg(['sum', 'count']).reset_index()
res_boat # 0 for no boat; people on the boat is nearly all alive
```
